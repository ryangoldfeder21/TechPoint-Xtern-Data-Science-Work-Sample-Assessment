```python
#searching for places w/keyword food truck and within 20 km of Indianapolis
places_result = gmaps_client.places_nearby(location = '39.791000, -86.148003', radius = 20000, keyword="food truck")

#creating an empty dataframe with columns my_fields
my_fields = ['name', 'formatted_address', 'rating', 'website', 'opening_hours']
address_list = []
name_list = []
rating_list = []
website_list = []
opening_hours_list = []
cuisine_list = []
foodTrucksDF = pd.DataFrame(columns=my_fields)

#loop 3 times to make dataframe of 60 food trucks (20 per API call)
for i in range(0, 3):
  for place in places_result['results']:
    my_place_id = place['place_id']
    my_fields = ['name', 'formatted_address', 'rating', 'website', 'opening_hours']

    #accessing specific details about the food truck(such as hours)
    place_details = (gmaps_client.place(place_id = my_place_id, fields = my_fields))

    #for name, address, rating, website, and hours, try-except statements handle cases with missing data
    try:
      name_list.append(place_details['result']['name'])
    except:
      name_list.append(np.nan)
    try:
      address_list.append(place_details['result']['formatted_address'])
    except:
      address_list.append(np.nan);
    try:
      rating_list.append(place_details['result']['rating'])
    except:
      rating_list.append(np.nan)
    try:
      website_list.append(place_details['result']['website'])
    except Exception:
      website_list.append(np.nan)
    try:
      opening_hours_list.append(place_details['result']['opening_hours']['weekday_text'])
    except Exception:
      opening_hours_list.append(np.nan)

    #yelp API to find cuisine of food truck, since Maps API does not include that as a category
    yelp_api_key = "5Z6Z-XBfSTPB_OhMx3P7sB9Q4uiEbZ3Qy6kd2w7m-i32iZqabdM-NzY7pL0xdpo-R9ds01i53JmyzM6vNiYQILM9RxJGdfcIAi9AOmPFzMshiGKhZc6yKBrsMRcuZXYx"
    ENDPOINT = 'https://api.yelp.com/v3/businesses/matches'
    HEADERS = {'authorization': 'bearer %s' % yelp_api_key}
    PARAMETERS = {'name': place_details['result']['name'], 'address1': place_details['result']['formatted_address'], 'city': 'Indianapolis','state': 'IN', 'country': 'US'}

    #use name and address from Maps API Search as a parameter in Yelp API search
    response = requests.get(url = ENDPOINT, params = PARAMETERS, headers = HEADERS)
    food_truck_data = response.json()
    try:
      truck = food_truck_data['businesses'][0]
      temp_endpoint = 'https://api.yelp.com/v3/businesses/{}'.format(truck['id'])
      response = requests.get(url = temp_endpoint, headers = HEADERS)
      cuisine = ""

      #categories variable contains list of possible cuisines
      for index in range (0, len(response.json()['categories'])):
        #categories variable also can include 'Food Trucks' and 'Caterers', which are not cuisines
        if (response.json()['categories'][index]['title'] != 'Food Trucks') & (response.json()['categories'][index]['title'] != 'Caterers'):
          #append to cuisine string in case list has multiple valid cuisines
          cuisine = cuisine + response.json()['categories'][index]['title'] + ', '
      if (cuisine == ""):
        #add null cuisine in case the only categories are 'Food Trucks' and/or 'Caterers'
        cuisine_list.append(np.nan)
      else:
        #delete comma at end of string
        cuisine = cuisine[0:len(cuisine) - 2]
        cuisine_list.append(cuisine)
    except:
      #add null cuisine in case of runtime error
      cuisine_list.append(np.nan)
  if i != 2:
    #call next page token to get next 20 results. time.sleep allows for the next page token to process in the system
    time.sleep(3)
    places_result = gmaps_client.places_nearby(page_token = places_result['next_page_token'])

#sets dataframe columns to recently created lists
foodTrucksDF['name'] = name_list
foodTrucksDF['formatted_address'] = address_list
foodTrucksDF['rating'] = rating_list
foodTrucksDF['website'] = website_list
foodTrucksDF['opening_hours'] = opening_hours_list
foodTrucksDF['cuisine'] = cuisine_list

#sorts rows of dataframe by rating in order to make it easier to pick the top rated food trucks to visit
foodTrucksDF = foodTrucksDF.sort_values(by='rating', na_position='last', ascending=0)

#allows for full opening hours string to print
pd.options.display.max_colwidth = 300
foodTrucksDF
```

```python
#creating fields and lists for travel plan dataframe
my_fields = ['Time', 'Name', 'Address', 'Cuisine', 'Travel Distance', 'Travel Time', 'Transportation Type']

#hardcoding names and times b/c I manually picked which food trucks to visit at each time
#chose food trucks with the highest rating that had opening hours listed
#scheduled based on which trucks were open/closed on Sundays and modified to reduce total travel time
#planned visits to 10 food trucks total (5 each day at 12, 3, 6, 7, and 10 PM)
name_list = ['big Jerry Indy food truck', 'The Latin Flavor', 'ClusterTruck - Broad Ripple', 'Kone Heads FoodTruck', 'Oasis Taco Truck',
             'King Gyro Food Truck- Emerson Avenue', 'El Buenavista Food Truck', 'KG Slider Station', 'El Rinconcito Salvadoreño food truck', 'The Night Owl Food Truck']
time_list = ['Sat. 12:00 PM' ,'Sat. 3:00 PM', 'Sat. 6:00 PM', 'Sat. 7:00 PM', 'Sat. 10:00 PM',
             'Sun. 12:00 PM', 'Sun. 3:00 PM', 'Sun. 6:00 PM', 'Sun. 7:00 PM', 'Sun. 10:00 PM']
address_list = []
cuisine_list = []
travel_dist_list = []
travel_time_list = []
transportation_type_list = []

#loop through to append null to travel distance, travel time, and transportation type
#these fields will be filled in later using the Maps / Directions API
for i in range (0, 10):
  travel_dist_list.append(np.nan)
  travel_time_list.append(np.nan)
  transportation_type_list.append(np.nan)
  #loops through Food Trucks Dataframe to find the given names and use the index to add the address and cuisine
  for j in range(0, len(foodTrucksDF)):
    if (foodTrucksDF['name'][j] == name_list[i]):
      address_list.append(foodTrucksDF['formatted_address'][j])
      cuisine_list.append(foodTrucksDF['cuisine'][j])

#creating the dataframe and setting the columns to the lists
foodTrucksPlanDF = pd.DataFrame(columns=my_fields)
foodTrucksPlanDF['Time'] = time_list
foodTrucksPlanDF['Name'] = name_list
foodTrucksPlanDF['Address'] = address_list
foodTrucksPlanDF['Cuisine'] = cuisine_list
foodTrucksPlanDF['Travel Distance'] = travel_dist_list
foodTrucksPlanDF['Travel Time'] = travel_time_list
foodTrucksPlanDF['Transportation Type'] = transportation_type_list
```

```python
#loop through food trucks on first day
for i in range(0, 5):
  if i == 0:
    #begin the day at IUPUI
    source = 'IUPUI'
  else:
    #start directions from previous food truck to next food truck
    source = foodTrucksPlanDF['Address'][i - 1]

  #destination is next/current food truck
  destination = foodTrucksPlanDF['Address'][i]

  #this variable holds lots of info about getting directions from source to destination
  my_directions = gmaps_client.directions(source, destination)

  #.loc function sets the Travel Distance, Travel Time, and Transportation Type to values in the my_directions json
  foodTrucksPlanDF.loc[i:i, ('Travel Distance')] = my_directions[0]['legs'][0]['distance']['text']
  foodTrucksPlanDF.loc[i:i, ('Travel Time')] = my_directions[0]['legs'][0]['duration']['text']
  foodTrucksPlanDF.loc[i:i, ('Transportation Type')] = my_directions[0]['legs'][0]['steps'][0]['travel_mode']

#loop through food trucks on second day
for i in range(5, 10):
  if i == 5:
    #begin the day at IUPUI
    source = 'IUPUI'
  else:
    #start directions from previous food truck to next food truck
    source = foodTrucksPlanDF['Address'][i - 1]

  #destination is next/current food truck
  destination = foodTrucksPlanDF['Address'][i]

  #this variable holds lots of info about getting directions from source to destination
  my_directions = gmaps_client.directions(source, destination)

  #.loc function sets the Travel Distance, Travel Time, and Transportation Type to values in the my_directions json
  foodTrucksPlanDF.loc[i:i, ('Travel Distance')] = my_directions[0]['legs'][0]['distance']['text']
  foodTrucksPlanDF.loc[i:i, ('Travel Time')] = my_directions[0]['legs'][0]['duration']['text']
  foodTrucksPlanDF.loc[i:i, ('Transportation Type')] = my_directions[0]['legs'][0]['steps'][0]['travel_mode']

#hardcoding cuisines of food trucks to visit because none of their cuisines were found by Yelp API
foodTrucksPlanDF.loc[0:0, ('Cuisine')] = 'Lunch Meats'
foodTrucksPlanDF.loc[1:1, ('Cuisine')] = 'Latin'
foodTrucksPlanDF.loc[2:2, ('Cuisine')] = 'Wide Range of Cuisines'
foodTrucksPlanDF.loc[3:3, ('Cuisine')] = 'American'
foodTrucksPlanDF.loc[4:4, ('Cuisine')] = 'Tacos'
foodTrucksPlanDF.loc[5:5, ('Cuisine')] = 'Greek'
foodTrucksPlanDF.loc[6:6, ('Cuisine')] = 'Mexican'
foodTrucksPlanDF.loc[7:7, ('Cuisine')] = 'American'
foodTrucksPlanDF.loc[8:8, ('Cuisine')] = 'Salvadoran, Mexican, Honduran'
foodTrucksPlanDF.loc[9:9, ('Cuisine')] = 'American'

#create CSV to download
foodTrucksPlanCSV = foodTrucksPlanDF.to_csv("Ryan Goldfeder - Finalized Food Trucks Plan.csv")
foodTrucksPlanDF
```

```python
#folium library to plot maps
import folium

#creating Saturday travel plan

#center of map is Indianapolis
indianapolis_coordinates = (39.791000, -86.148003)
saturdayMap = folium.Map(indianapolis_coordinates, zoom_start=11)
lat_coord_list = []
lon_coord_list = []

#this function takes a place and converts it to coordinates
geocode_result = gmaps_client.geocode("IUPUI")

#adding coordinates to lists to loop through lter
lat_coord = geocode_result[0]["geometry"]["location"]["lat"]
lon_coord = geocode_result[0]["geometry"]["location"]["lng"]
lat_coord_list += [lat_coord]
lon_coord_list += [lon_coord]
#create circle at current coordinates
folium.CircleMarker([lat_coord, lon_coord], radius=5, fill_color="blue").add_to(saturdayMap)
#add popup w/text to circle
saturdayMap.add_child(folium.Marker(location= [lat_coord, lon_coord], popup="IUPUI", icon=folium.Icon(color='black')))

for i in range(0, 5):

  #geocoding each address of food truck
  geocode_result = gmaps_client.geocode(foodTrucksPlanDF['Address'][i])

  lat_coord = geocode_result[0]["geometry"]["location"]["lat"]
  lon_coord = geocode_result[0]["geometry"]["location"]["lng"]
  lat_coord_list += [lat_coord]
  lon_coord_list += [lon_coord]

  #string with food truck name and time
  str = foodTrucksPlanDF['Name'][i] + ": " + foodTrucksPlanDF['Time'][i]

  #add circle and popup/w food truck info
  folium.CircleMarker([lat_coord, lon_coord], radius=5, fill_color="blue").add_to(saturdayMap)
  saturdayMap.add_child(folium.Marker(location= [lat_coord, lon_coord], popup=str, icon=folium.Icon(color='black')))

#loop through to add lines connecting each location
for i in range(0, 5):
  folium.PolyLine([[lat_coord_list[i], lon_coord_list[i]],
                     [lat_coord_list[i+1], lon_coord_list[i+1]]]).add_to(saturdayMap)
saturdayMap
```
